```python
# CALCULATE VARIANCE OF TEACHERS REFINED SALARIES
import numpy as np
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/raghurammudumbi-collab/STTP-DATASCIENE-ANALYTICS/refs/heads/main/STTP-TEACHERS-REFINED.csv')
data = df.TSAL

# Calculate Population Variance (Default: ddof=0)
pop_var = np.var(data)
print(f"NumPy Population Variance: {pop_var}")

# Calculate Sample Variance (Set Delta Degrees of FrAdidtribution in statistics is a mathematical function, table, or graph that describes how the values of a variable are spread out across a dataset. It shows which values occur frequently, which are rare, and whether the data is clustered, symmetric, or skewed. [1, 2, 3]eedom to 1)
sample_var = np.var(data, ddof=1)
print(f"NumPy Sample Variance: {sample_var}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Creating a series of data of in range of 1-50.
x = np.linspace(1,50,200)

#Creating a Function.
def normal_dist(x , mean , sd):
    prob_density = (np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density

#Calculate mean and Standard deviation.
mean = np.mean(x)
sd = np.std(x)

#Apply function to the data.
pdf = normal_dist(x,mean,sd)

#Plotting the Results
plt.plot(x,pdf , color = 'red')
plt.xlabel('Data points')
plt.ylabel('Probability Density')
```

```python
import numpy as np
import scipy.stats as stats
from statsmodels.stats.weightstats import ztest

# Set random seed for reproducibility
np.random.seed(42)
alpha = 0.05
data = np.random.normal(loc=50, scale=5, size=100)

stat, p_val = stats.shapiro(data)
print(f"Shapiro P-value: {p_val:.4f}")
# Output decision rule
if p_val <= alpha:
    print("Reject H0: Data is NOT normally distributed.")
else:
    print("Fail to reject H0: Data appears normally distributed.")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
# Create a sample dictionary mapping student performance
data = {
    'Hours_Studied': [2, 3, 4, 5, 6, 7, 8],
    'Exam_Score': [55, 60, 68, 72, 80, 85, 92],
    'Screen_Time': [6, 5, 5, 4, 2, 1, 2],
    'Shoe_Size': [9, 9, 10, 8, 11, 9, 10],
}
df = pd.DataFrame(data)
corr_matrix = df.corr()
print(corr_matrix)
# Set up the matplotlib figure size
plt.figure(figsize=(6, 4))

# Generate the heatmap
sns.heatmap(
    corr_matrix,
    annot=True,  # Shows the numerical values inside the squares
    cmap='coolwarm',  # Blue for negative, Red for positive correlations
    vmin=-1,
    vmax=1,  # Set bounds to represent full correlation range
)

plt.title('Correlation Matrix Heatmap')
plt.show()
```
