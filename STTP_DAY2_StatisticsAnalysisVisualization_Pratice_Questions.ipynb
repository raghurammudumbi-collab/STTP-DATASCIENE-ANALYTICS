```python
import numpy as np
import pandas as pd
import statistics as stats

data = [4, 8, 15, 16, 23, 42, 42]

# Central Tendency
mean_val = np.mean(data)       # 21.42
median_val = np.median(data)   # 16.0
mode_val = stats.mode(data)    # 42

# Measures of Dispersion (Spread)
variance = np.var(data, ddof=1)  # Sample variance (Divided by N-1)
std_dev = np.std(data, ddof=1)   # Sample standard deviation
data_range = np.ptp(data)        # Peak-to-peak (Max - Min)
```

```python
import matplotlib.pyplot as plt

# Prepare mock data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
revenue = [12000, 15000, 14000, 19000, 22000]

# Build the layout
plt.figure(figsize=(8, 4))
plt.plot(months, revenue, marker='o', color='#1f77b4', linewidth=2)

# Structure and clarity
plt.title('Company Revenue Over Time', fontsize=14, fontweight='bold')
plt.xlabel('Month', fontsize=11)
plt.ylabel('Revenue ($)', fontsize=11)
plt.grid(True, linestyle='--', alpha=0.6)

# Display result
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a classic built-in dataset
tips = sns.load_dataset("tips")

# Set sophisticated style defaults
sns.set_theme(style="whitegrid")

# Create a multi-dimensional scatter plot
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    style="time",
    size="size"
)

plt.title('Analysis of Restaurant Tips', fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 16]

# Customizing line properties
plt.plot(x, y,
         color='green',       # Set line color
         linestyle='--',      # Dashed line style
         linewidth=2.5,       # Thickness of the line
         marker='o',          # Circular markers at data points
         markersize=8,        # Size of the markers
         label='Temperature') # Label for the legend

# Enhancing the visual presentation
plt.title("Customized Trend Chart")
plt.xlabel("Days")
plt.ylabel("Value")
plt.grid(True)                # Turn on the background grid
plt.legend()                  # Show the legend box

plt.show()
```

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
product_a_sales = [100, 120, 170, 140, 210]
product_b_sales = [90, 110, 130, 150, 180]

# Plot each line independently
plt.plot(months, product_a_sales, color='blue', marker='s', label='Product A')
plt.plot(months, product_b_sales, color='red', marker='^', label='Product B')

# Graph details
plt.title("Monthly Product Sales Comparison")
plt.xlabel("Months")
plt.ylabel("Sales (Units)")
plt.legend()                  # Categorically distinguishes the lines
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Q1', 'Q2', 'Q3', 'Q4']
sales_2025 = [20, 35, 30, 35]
sales_2026 = [25, 32, 34, 45]

# Set the positions and width of the bars
x = np.arange(len(categories))
width = 0.35  # Thickness of each bar

# Plot both datasets with offsets
plt.bar(x - width/2, sales_2025, width, label='2025', color='royalblue')
plt.bar(x + width/2, sales_2026, width, label='2026', color='orange')

# Format axes
plt.xlabel('Quarters')
plt.ylabel('Revenue (Millions)')
plt.title('Year-over-Year Quarterly Revenue')
plt.xticks(x, categories)  # Replace numeric indices with text labels
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['Store A', 'Store B', 'Store C']
online_sales = [50, 80, 45]
in_store_sales = [70, 40, 55]

# Plot base layer
plt.bar(categories, online_sales, label='Online', color='teal')

# Plot top layer (using 'bottom' to stack)
plt.bar(categories, in_store_sales, bottom=online_sales, label='In-Store', color='coral')

plt.ylabel('Total Orders')
plt.title('Sales Channel Breakdown')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Dataset
sizes = [40, 30, 20, 10]
labels = ['Python', 'Java', 'C++', 'Ruby']
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

# Explode the 1st slice (Python) out by 0.1 units
explode = (0.1, 0, 0, 0)

plt.figure(figsize=(6, 6))

plt.pie(
    sizes,
    explode=explode,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',     # Formats percentage to 1 decimal place
    shadow=True,           # Adds a 3D drop shadow effect
    startangle=140         # Rotates the start point counter-clockwise
)

plt.title("Programming Language Popularity")
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [40, 30, 20, 10]
labels = ['Python', 'Java', 'C++', 'Ruby']

# Main pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', pctdistance=0.85)

# Create a circle for the center hole
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Donut Chart Example")
plt.show()
```
