```python
# LOAD TEACHERS_RAW DATA and Do one-hot encoding

import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/raghurammudumbi-collab/STTP-DATASCIENE-ANALYTICS/refs/heads/main/STTP-TEACHERS-RAW.csv')

# df = pd.DataFrame(data)

print("Original Data:")

print(df)

encoded_df = pd.get_dummies(
    df,
    columns=['TSUB'],
    drop_first=True
)

print("\nOne-Hot Encoded Data:")
print(encoded_df)
```

```python
## To check for duplicates in TEACHERS_RAW Data
import pandas as pd

# Load the CSV data
df = pd.read_csv('https://raw.githubusercontent.com/raghurammudumbi-collab/STTP-DATASCIENE-ANALYTICS/refs/heads/main/STTP-TEACHERS-RAW.csv')

# Get a boolean mask of duplicate rows (True means it is a duplicate)
duplicate_mask = df.duplicated()

# Print the total number of duplicate rows
print(f"Total duplicate rows: {duplicate_mask.sum()}")

# Display the actual duplicate rows (keeps the first occurrence)
duplicate_rows = df[duplicate_mask]
print(duplicate_rows)
```

```python
import csv

# Define the data to be written to the CSV file
data = [
    ['Name', 'Age', 'City'],
    ['Alice', 25, 'New York'],
    ['Bob', 30, 'Los Angeles'],
    ['Charlie', 35, 'Chicago']
]

# Specify the file name
filename = 'C:\\raghutest.csv'

# Write data to the CSV file with a pipe delimiter
with open(filename, 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile, delimiter='|')
    csvwriter.writerows(data)

print(f"Data has been written to {filename}")
```
