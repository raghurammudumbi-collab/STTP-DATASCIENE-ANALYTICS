```python
import numpy as np

a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

# Element-wise addition
print(a + b)
# Output: [11 22 33 44]

# Element-wise multiplication
print(b * 2)
# Output: [2 4 6 8]

# Squaring all elements
print(b ** 2)
# Output: [1 4 9 16]
```

```python
import numpy as np

arr = np.array([10, 15, 20, 25, 30])

# Slice from index 1 to 4 (exclusive)
print(arr[1:4])
# Output: [15 20 25]

# Filter elements greater than 20
condition = arr > 20
print(arr[condition])
# Output: [25 30]
```

```python
import numpy as np

matrix = np.array([[1, 2], [3, 4]])

# Sum of all elements
print(np.sum(matrix))       # Output: 10

# Maximum value
print(np.max(matrix))       # Output: 4

# Average (Mean) value
print(np.mean(matrix))      # Output: 2.5

# Column-wise sum (Axis 0)
print(np.sum(matrix, axis=0)) # Output: [4 6]
```

```python
import numpy as np

matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[2, 0], [1, 2]])

# Dot product of matrices
product = np.dot(matrix_a, matrix_b)
# Output:
# [[ 4  4]
#  [10  8]]
```
