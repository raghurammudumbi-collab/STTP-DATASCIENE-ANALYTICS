```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Create a hypothetical Teacher/Student database
data = {
    'Teacher_Name': ['Sarah', 'John', 'Emily', 'Michael', 'David'],
    'Class_Size': [22, 35, 18, 40, 28],
    'Average_Test_Score': [88, 72, 92, 68, 85]
}
df = pd.DataFrame(data)

# 2. Visualize using a scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(df['Class_Size'], df['Average_Test_Score'], color='blue', s=100)

# 3. Add titles and labels
plt.title('Impact of Class Size on Average Test Scores')
plt.xlabel('Number of Students in Class')
plt.ylabel('Average Test Score (out of 100)')
plt.grid(True)

# 4. Display the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Create a hypothetical Student database linked to Teachers
data = {
    'Student_Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank', 'Grace', 'Heidi'],
    'Teacher': ['Mr. Smith', 'Mr. Smith', 'Mrs. Davis', 'Mrs. Davis', 'Mr. Smith', 'Mrs. Davis', 'Mr. Smith', 'Mrs. Davis'],
    'Exam_Score': [85, 90, 78, 92, 88, 81, 95, 89]
}
df = pd.DataFrame(data)

# 2. Visualize distributions using a Seaborn box plot
plt.figure(figsize=(8, 5))
sns.boxplot(x='Teacher', y='Exam_Score', data=df, palette='Set2')

# 3. Add titles and labels
plt.title('Student Score Distribution by Teacher')
plt.xlabel('Teacher')
plt.ylabel('Exam Score')

# 4. Display the plot
plt.show()
```

```python
import pandas as pd

# Base dataset from Example 1
data_teachers = {
    'Teacher_Name': ['Sarah', 'John', 'Emily', 'Michael', 'David'],
    'Class_Size': [15, 30, 22, 18, 35],
    'Average_Test_Score': [92, 75, 88, 81, 70]
}
df_teachers = pd.DataFrame(data_teachers)

# Apply multiple filtering conditions using the & (AND) operator
# Each condition must be wrapped in parentheses
filtered_classes = df_teachers[(df_teachers['Class_Size'] < 25) & (df_teachers['Average_Test_Score'] >= 85)]

print(filtered_classes)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load combined teacher and student data
data = {
    "Teacher_ID": [101, 102, 103, 104],
    "Experience_Years": [2, 10, 5, 15],
    "Class_Average_Score": [75, 88, 80, 95],
}
df = pd.DataFrame(data)

# Calculate correlation
correlation = df["Experience_Years"].corr(df["Class_Average_Score"])
print(f"Correlation: {correlation}")

# Create scatter plot
sns.scatterplot(
    x="Experience_Years", y="Class_Average_Score", data=df, s=100
)
plt.title("Teacher Experience vs. Class Average")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
import pandas as pd

# Load training data
data = {
    "Study_Hours": [10, 25, 15, 35, 5, 30],
    "Attendance_Pct": [80, 95, 85, 99, 60, 92],
    "Passed": [0, 1, 0, 1, 0, 1],  # 0 = Fail, 1 = Pass
}
df = pd.DataFrame(data)

# Train the model
X = df[["Study_Hours", "Attendance_Pct"]]
y = df["Passed"]
model = LogisticRegression()
model.fit(X, y)

# Predict for a new student with 20 hours of study and 90% attendance
prediction = model.predict([[20, 90]])
print(f"Predicted Pass/Fail (1=Pass, 0=Fail): {prediction[0]}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data
data = {
    "Teacher_ID": ["Mr. Smith", "Ms. Davis", "Mr. Jones"],
    "Class_Size": [15, 30, 22],
    "Average_Grade": [88, 79, 84],
}
df = pd.DataFrame(data)

# Create a bar chart
plt.bar(df["Teacher_ID"], df["Average_Grade"], color=["salmon", "gold", "teal"])
plt.title("Average Grade by Teacher (Class Size)")
plt.xlabel("Teacher")
plt.ylabel("Average Grade")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load time-series data
data = {
    "Date": ["2026-05-01", "2026-05-08", "2026-05-15", "2026-05-22"],
    "Submissions": [45, 48, 42, 35],
}
df = pd.DataFrame(data)
df["Date"] = pd.to_datetime(df["Date"])

# Create a line chart
plt.plot(
    df["Date"], df["Submissions"], marker="o", color="green", linestyle="-"
)
plt.title("Homework Submissions Trend (May 2026)")
plt.xlabel("Date")
plt.ylabel("Number of Submissions")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
